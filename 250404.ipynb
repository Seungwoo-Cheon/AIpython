```python
type = input("온도 종류(섭씨 또는 화씨): ")
temp = int(input(f"{type} 온도: "))

if type == "섭씨":
  f = temp*(9/5)+32
  print(f"화씨 온도: {f}")
  print("화씨 온도: %.2f" % (f))
else:
  c = (temp-32)*(5/9)
  print(f"섭씨 온도: {c}")
  print("섭씨 온도: %.2f" % (c))
```

```python
import random

print("1부터 100까지의 수 하나를 정했습니다.\n이 숫자를 맞춰보세요.")

a = random.randint(1,100)
t = 0

while True:
  b = int(input("1~100 숫자 입력: "))
  t += 1
  if a == b:
    break
  elif a > b:
    print(f"{b}보다 큽니다.")
  else:
    print(f"{b}보다 작습니다.")

print(f"정답입니다! {t}회 만에 맞췄습니다.")
```

```python
import random

n = int(input())

l = []
t = 0

while True:
  t += 1
  if t < (n+1):
    l.append(random.randint(1,45))
  else:
    break

print(l)
```

```python
l = [85, 87, 78, 90, 92]
s = 0
n = 0

for i in l: #반복은 리스트 길이 만큼
  s += i
  n += 1

mean = s/n
print(mean)
```

```python
l = [85, 87, 78, 90, 92]

mean = sum(l)/len(l)

print(mean)
```
